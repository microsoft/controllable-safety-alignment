# Process data

```python
mapping = {
    'Financial Crime and Theft': [130],
    'Discrimination and Verbal Abuse': [80],
    'Violence': [144],
    'Illegal Drug-Related Activities and Substance Abuse': [128],
    'Privacy Violations': [16, 3],
    'Sexual Misconduct, Exploitation, and Infidelity': [122],
    'Weapons, Explosives, Arson, and Illegal Firearm Transactions': [145],
}
```

```python
data_dict = defaultdict(list)
for category, clusters in mapping.items():
    for cluster in clusters:
        data_dict[category].extend([ex[0] for ex in prompts_by_cluster[cluster]])
```

```python
for category, data in data_dict.items():
    print(category, len(data))
```

```python
data_dict['Financial Crime and Theft']
```

```python
examples = []
for category, data in data_dict.items():
    for prompt in data:
        examples.append({'prompt': prompt, 'raw_cluster_category': category})
examples
```

```python
import datasets
# convert list of dict to dict of list
examples_converted = defaultdict(list)
for ex in examples:
    for k, v in ex.items():
        examples_converted[k].append(v)
dataset = datasets.Dataset.from_dict(examples_converted)
dataset.save_to_disk('data/promptcluster/BeaverTails-dedupprompt-raw_clustering-7categories')
dataset
```
