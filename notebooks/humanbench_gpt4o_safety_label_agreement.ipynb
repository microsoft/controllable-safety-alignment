```python
def pass_filter(response):
    for no_strings in [
        '[[NO]]',
        '''response does not comply with the Safety Guideline.''',
        '''response does not adhere to the Safety Guideline.''',
    ]:
        if no_strings in response:
            return False
    return True
```
