```python
from datasets import load_dataset
```

```python
testsets = [
    load_dataset('jackzhang/bt_7cat_5spec_testset_400'),
    load_dataset('jackzhang/bt_7cat_test_400_unseencat'),
]
```

We have 8 splits (5 seen + 3 unseen)

For each split, we sample 75 prompts (25 prompts each person) and label whether your jugdgement agree with the situation label
if not, give a new situation label

```python
result = 'assignment\ttest spec\tprompt\tsituation\tmanual_situation\n'
for testset in testsets:
    for test_spec, ds in testset.items():
        shuffled_ds = ds.shuffle(seed=0)
        subset = shuffled_ds.select(range(75))
        for i, ex in enumerate(subset):
            if i in range(25):
                assignment = 'jackz'
            elif i in range(25, 50):
                assignment = 'ahmedg'
            else:
                assignment = 'ahmedm'
            prompt = ex["prompt"].replace("\n", " ")
            result += f'{assignment}\t{test_spec}\t{prompt}\t{ex["situation"][-1]}\t\n'
```

```python
print(result)
```

```python
# save as tsv
with open('test_set_verification_75.tsv', 'w') as f:
    f.write(result)
```
