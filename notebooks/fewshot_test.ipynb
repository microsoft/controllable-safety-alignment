```python
# from datasets import load_dataset
from datasets import load_from_disk
from collections import defaultdict
from src.produce_metrics import cat_str_to_list
import pickle
```

```python
# ds = load_dataset('jackzhang/V5-bt-wg-addr_imp-train')['train']
ds = load_from_disk('data/paired/V5-bt+wg-addr_imp/train_7cat')
```

```python
ds
```

```python
speci2prompt = defaultdict(set)
speci2promptresponse = defaultdict(list)

for ex in ds:
    current_speci = tuple(sorted(cat_str_to_list(ex['spec_i'])))
    if ex['prompt'] in speci2prompt[current_speci]:
        # pass
        continue
    speci2prompt[current_speci].add(ex['prompt'])
    speci2promptresponse[current_speci].append((ex['prompt'], ex['chosen_response']))
for k, v in speci2promptresponse.items():
    print(k, len(v))
print(sum(len(v) for v in speci2promptresponse.values()))
```

```python
# save to disk as pickle
with open('data/speci2pr/V5_train_7cat.pkl', 'wb') as f:
    pickle.dump(speci2promptresponse, f)
```
