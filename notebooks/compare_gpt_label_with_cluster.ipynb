````python
def str_to_json(s, single_line=True):
    s = s.strip()
    json_portion = s.split('\n')[-1] if single_line else s
    if not s.startswith('{'):
        json_start = s.find('{')
        json_portion = s[json_start:]
        json_portion.replace('```', '')
    return json.loads(json_portion.replace('Final Answer: ', '').replace("'", '"'))
````
